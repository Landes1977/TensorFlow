```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Data   : http://users.stat.ufl.edu/~winner/data/brainhead.dat

Labels : http://users.stat.ufl.edu/~winner/data/brainhead.txt

**Columns**
- Gender
  * 1=Male 
  * 2=Female
- Age Range
  * 1=20-46
  * 2=46+
- Head size (cm^3)
- Brain weight (grams)

```python
fname = "http://users.stat.ufl.edu/~winner/data/brainhead.dat"
df = pd.DataFrame(np.loadtxt(fname),
                  columns=['Gender', 'AgeRange', 'HeadSize', 'BrainWeight'])
```

```python
df.head()
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import preprocessing

reg = linear_model.LinearRegression()
reg.fit(df[['Gender', 'AgeRange', 'HeadSize']], df['BrainWeight'])

print('***** Coefficents *****')
print('Gender: {:.2f}'.format(reg.coef_[0]) + '\n' +
      'AgeRange: {:.2f}'.format(reg.coef_[1]) + '\n' +
      'HeadSize: {:.2f}'.format(reg.coef_[2]) + '\n'
     )

print('***** Interpretation *****')
print('On average, female adult brains weight {:.2f}'.format(np.abs(reg.coef_[0])) + ' grams less than male adult brains.')
print('On average, older adults (46+) brains weight {:.2f}'.format(np.abs(reg.coef_[1])) + ' grams less than younger adults (20-45).')
print('On average, an adults brain weights an additional {:.2f}'.format(reg.coef_[2]) + ' grams for each additional cubic centimeter.')
```

```python
x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)


df2 = pd.DataFrame(x_scaled, 
                   columns=['Gender', 'AgeRange', 'HeadSize', 'BrainWeight'])

reg.fit(df2[['Gender', 'AgeRange', 'HeadSize']], df2['BrainWeight'])
reg.coef_
```
